```python
import random
from random_word import RandomWords

print('Time To Play Hangman!')
r = RandomWords()
word_list= r.get_random_words()
secret = random.choice(word_list)
guesses = ''
already_guessed = []

turns = 6

while turns > 0:
  missed = 0

  for letter in secret:
    if letter in guesses:
      print(letter)
    else:
      print('_')
      missed += 1

  if missed == 0:
    print('You win!')
    print('The answer is:', secret)
    break

  guess = input('guess a letter: ')
  guesses += guess
  already_guessed.extend([guess])
  print('Already guesses: ' ,already_guessed)  
    
  if len(guess) !=1:
            print("!!Pick One Letter At Same Time!!")
        
  elif guess not in 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwyz- ':
            print("!!Please Only Pick Letters!!")   

  elif guess not in secret:
    turns -= 1
    print('Nope!')
    print('You have', turns , 'more turns')
    if turns == 5:   
        print("________      ")
        print("|      |      ")
        print("|             ")
        print("|             ")
        print("|             ")
        print("|             ")
    if turns == 4: 
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|             ")
        print("|             ")
        print("|             ")
    if turns == 3: 
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|      ")
        print("|             ")
        print("|             ")
    if turns == 2: 
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|\     ")
        print("|             ")
        print("|             ")
    if turns == 1: 
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|\     ")
        print("|     /       ")
        print("|             ")
    if turns == 0:
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|\     ")
        print("|     / \     ")
        print("|             ")
        print('The answer is:', secret)
```
